```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('california_housing.csv')
df
```

```python
from sklearn.datasets import fetch_california_housing
california_housing = fetch_california_housing(as_frame=True)
df_ca = california_housing.frame
df_ca
```

```python
df_ca.to_csv("california_housing.csv", index=False)
import pandas as pd
df_ca2 = pd.read_csv("california_housing.csv")
df_ca2.head()
```

```python
#creating X as features/independent variables and y as target/dependent variable
X = df_ca[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'AveOccup']]
y = df_ca['MedHouseVal']
```

```python
#Create train and test data
#Split the data into 80% training dataset and 20% test dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=123)
```

```python
#Perform Linear Regression
from sklearn.linear_model import LinearRegression 
from sklearn import metrics
from sklearn.metrics import mean_squared_error

reg = LinearRegression() 
  
reg.fit(X_train, y_train) 

print(reg.coef_)
print(reg.intercept_)
```

```python
#root mean squared error (RMSE) from Linear Regression
y_pred = reg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
x = np.arange(50)

plt.rcParams["figure.figsize"] = (12,4)
plt.plot(x, y_pred[:50], c = "red")
plt.plot(x, y_test[:50], c = "blue")
plt.show()
```
